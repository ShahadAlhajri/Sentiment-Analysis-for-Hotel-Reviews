```python
# 1-Import necessary libraries
import pandas as pd
```

```python
# 2-Load the dataset and display rating counts
df = pd.read_csv("/content/tripadvisor_hotel_reviews.csv")
df
```

```python
# 3-Select, balance, and preprocess the reviews

df_neg = df[df["Rating"] < 3].reset_index(drop=True)
df_pos = df[df["Rating"] == 5].reset_index(drop=True)

# Balance the dataset by limiting the number of positive reviews to match the number of negative reviews
df_pos_balanced = df_pos.loc[:len(df_neg)-1]

# Concatenate the negative and balanced positive reviews into a single DataFrame and reset the index
df_all = pd.concat([df_neg, df_pos_balanced], axis=0).reset_index(drop=True)

# 'Sentiment' is a binary target variable derived from 'Rating'
df_all['Sentiment'] = df_all['Rating'].apply(lambda x: 'positive' if x > 2 else 'negative')

# Remove the 'Rating' column from df_all
df_all = df_all.drop('Rating', axis=1)

df_all
```

```python
# 4-Save the DataFrame to a CSV file
df_all.to_csv('Hotel_reviews.csv', index=False)
```
